```python
import pandas as pd
import numpy as np
```

```python
census_df = pd.read_csv("1222_aggregated_census_employment_data - 1222_aggregated_census_employment_data.csv")
census_df.head()
```

```python
jobs = ["Manufacturing and Warehousing", "Retail", "Service", "Office", "Institutional", "Community and Entertainment", "Grand Total"]
census_df.drop(jobs, axis=1, inplace=True)
census_df = census_df.drop(columns=[col for col in census_df.columns if col.lower().endswith('total')])
census_df.drop(0, axis=0, inplace=True)
census_df = census_df[census_df["CTUID"].between(5350001, 5350379)]
census_df['CTUID'] = census_df['CTUID'].apply(lambda x: f"{float(x):.2f}")
census_df
```

```python
metric_df = pd.read_csv("total_Metric_Table_Ontario_Line_Update.csv")
metric_df = metric_df[metric_df["Metric_Name"] == "Job Access"]
metric_df['CTUID'] = metric_df['CTUID'].apply(lambda x: f"{float(x):.2f}")
metric_df
```

```python
metric_col = metric_df.columns.to_list()
metric_col
```

```python
merged_df = pd.merge(metric_df, census_df, how ="left", on ="CTUID")
merged_df = merged_df.fillna(0)
merged_df
```

```python
census_col = census_df.columns.to_list()[1:]
merged_df=merged_df.melt(id_vars=metric_col, value_vars=census_col, var_name="Attribute", value_name="Census_Value")
```

```python
merged_df["Attribute(groups)"] = merged_df['Attribute'].apply(lambda x: x.split("_")[0])
merged_df["Attribute(gender)"] = merged_df["Attribute"].apply(lambda x: x.split("_")[1])
merged_df
```

```python
merged_df_2 = merged_df.copy()
```

```python
merged_df_2['Attribute_Sum'] = merged_df_2.groupby(["CTUID", "Attribute(groups)", "Attribute(gender)",
                                                    "Metric_Type", "Metric_Name", "Travel_Time_Threshold", "Category",
                                                    "Before_After_Benefit"])['Census_Value'].transform('sum')
```

```python
merged_df_2["Pop_Prop"] = merged_df_2["Census_Value"]/merged_df_2["Attribute_Sum"]
```

```python
df = merged_df_2
```

```python
# Read the CSV
population_data = df[
    (df['Neighbourhoods'] == 'South Riverdale') & 
    (df['Attribute'] == 'Population_2021')
][['CTUID', 'Census_Value']].drop_duplicates()

south_riverdale_benefits = df[
    (df['Neighbourhoods'] == 'South Riverdale') &
    (df['Before_After_Benefit'] == 'benefit') &
    (df['Category'] == 'Grand Total')
]

# Filter for South Riverdale and benefit
south_riverdale_benefits = south_riverdale_benefits.merge(
    population_data, 
    on='CTUID', 
    suffixes=('', '_pop')
)

south_riverdale_benefits = south_riverdale_benefits[south_riverdale_benefits['Pop_Prop'] != 0]
south_riverdale_benefits["Census_Value_pop"] = south_riverdale_benefits[["Census_Value_pop", "Attribute_Sum"]].values.max(1)
south_riverdale_benefits['Metric_Per_Capita'] = south_riverdale_benefits['Value'] / ((south_riverdale_benefits["Pop_Prop"] * south_riverdale_benefits["Census_Value_pop"]))

sorted_data = south_riverdale_benefits.sort_values(by=["Metric_Per_Capita", 'CTUID'], ascending=False)

#sorted_data = sorted_data.drop(["Census_Value_pop", "Proportion_Value"])
#Display the filtered dataframe
print(sorted_data)

#Optional: Save to a new CSV if needed
sorted_data.to_csv('south_riverdale_benefits_census.csv', index=False)
```
